```python
raw = pd.read_csv("Churn_Modelling.csv")
raw
```

```python
data = raw.copy()
data
```

```python
x = data.drop(["RowNumber", "CustomerId", "Surname", "Exited"], axis = 1)
y = data["Exited"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le
```

```python
x["Geography"] = le.fit_transform(x["Geography"])
```

```python
x["Gender"] = le.fit_transform(x["Gender"] )
```

```python
x
```

```python
x = pd.get_dummies(x, drop_first = True, columns = ["Geography"])
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 420)
```

```python
print("x_train shape: {}\ny_train shape: {}\nx_test shape: {}\ny_test shape: {}".format(x_train.shape, y_train.shape, x_test.shape, y_test.shape))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
y_train.size
```

```python
y_test.size
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units = 6, activation = "relu", input_dim = 11))
model.add(tf.keras.layers.Dense(units = 6, activation = "relu"))

model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))
```

```python
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = "accuracy")
model.summary()
```

```python
y_train.shape, x_train.shape, y
```

```python
history =model.fit(x_train, y_train, epochs = 15, validation_data = (x_test, y_test))
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
```

```python
print("test_loss: {}\ntest_accuracy: {}".format(test_loss *100, test_accuracy * 100))
```

```python
y_pred = (model.predict(x_test)>0.5).astype(np.int32)
```

```python
y_test = np.array(y_test)
```

```python
print(y_pred[4], y_test[[4]])
```
