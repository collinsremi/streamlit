```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print("x_train.shape: {}\ny_train.shape: {}\nx_test.shape: {}\ny_test.shape: {}".format(x_train.shape, y_train.shape, x_test.shape, y_test.shape))
```

```python
from random import randint
```

```python
plt.figure(figsize = (20, 10), dpi = 120)
for i in range(1, 66):
  plt.subplot(5, 13, i)
  img = randint(1, 50000)
  plt.imshow(x_train[img])
  plt.title(str(y_train[img]));
```

```python
plt.imshow(x_train[334])
plt.colorbar();
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
plt.imshow(x_train[334])
plt.colorbar();
```

```python
x_train.max()
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 2, padding = "same", activation = "relu", input_shape = (32, 32, 3)))
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 2, padding = "valid", activation = "relu"))
```

```python
model.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2, padding = "valid"))
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 2, padding = "valid", activation = "relu"))
```

```python
model.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2, padding = "valid"))
```

```python
model.add(tf.keras.layers.Dropout(0.5))
```

```python
model.add(tf.keras.layers.Flatten())
```

```python
model.add(tf.keras.layers.Dense(units = 140, activation = "relu"))
```

```python
model.add(tf.keras.layers.Dense(units = 80, activation = "relu"))
```

```python
model.add(tf.keras.layers.Dropout(0.5))
```

```python
model.add(tf.keras.layers.Dense(units = 10, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = "accuracy")
```

```python
history = model.fit(x_train, y_train, epochs = 15, validation_data = (x_test, y_test), batch_size = 20)
```

```python
def diagram(epochs, hist):
  plt.plot(epochs, hist.history["accuracy"], label = "accuracy")
  plt.plot(epochs, hist.history["val_accuracy"], label = "val_accuracy")
  plt.legend(shadow = True, loc = "lower right")
  plt.show()
  plt.plot(epochs, hist.history["loss"], label = "loss")
  plt.plot(epochs, hist.history["val_loss"], label = "val_loss")
  plt.legend(shadow = True, loc = "upper right")
  plt.show()
```

```python
diagram(range(1, 16), history)
```
