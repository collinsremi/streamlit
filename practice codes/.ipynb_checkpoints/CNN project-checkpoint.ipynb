```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

# flatling the data


```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
plt.imshow(x_train[233])
plt.colorbar();
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, padding = "same", activation = "relu", input_shape = (32, 32, 3)))
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, padding = "same", activation = "relu"))
```

```python
model.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2, padding = "valid"))
```

```python
model.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, padding = "same", activation = "relu"))
```

```python
model.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2, padding = "valid"))
```

```python
model.add(tf.keras.layers.Dropout(0.2))
```

```python
model.add(tf.keras.layers.Flatten())
```

```python
model.add(tf.keras.layers.Dense(units = 120, activation = "relu"))
```

```python
model.add(tf.keras.layers.Dense(units = 120, activation = "relu"))
```

```python
model.add(tf.keras.layers.Dense(units = 10, activation = "softmax"))
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = "sparse_categorical_accuracy")
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs = 10, validation_data = (x_test, y_test), batch_size= 20)
```

```python
y_pred = np.argmax(model.predict(x_test), axis = -1)
```

```python
y_pred[2], y_test[2]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
con_ma = confusion_matrix(y_pred, y_test)
```

```python
con_ma
```

```python
acc_sc = accuracy_score(y_pred, y_test)
```

```python
acc_sc
```

```python
epochs = range(1, 11)
hist = history.history
plt.plot(epochs, hist["sparse_categorical_accuracy"], label = "s_c_a")
plt.plot(epochs, hist["val_sparse_categorical_accuracy"], label = "val_s_c_a")
plt.legend(shadow = True);
```

```python
epochs = range(1, 11)
hist = history.history
plt.plot(epochs, hist["loss"], label = "loss")
plt.plot(epochs, hist["val_loss"], label = "val_l")
plt.legend(shadow = True);
```
